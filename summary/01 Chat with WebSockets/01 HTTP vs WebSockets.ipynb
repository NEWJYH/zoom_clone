# real-time(실시간)을 하기전 어떤 것이 실시간을 가능하게 해주는지?



## HTTP & WebSocket

둘다 protocol(프로토콜)

## HTTP ??

모든 서버들이 작동하는 방식

유저 request, 서버 response 

인터넷 전체가 http를 기반으로 만들어짐

http에서 중요한 점은 stateless 
- backend가 유저를 기억하지 못한다는 점
- user와 backend 사이에 아무런 연결이 없음
- request와 response과정 뒤에 backend는 유저를 잊어버림
- 서버는 오직 request를 받을 때만 답장을 해줌
- real-time으로 일어나지 않음




## WebSocket
chat, notification 같은 real-time을 만들수 있음

- webSocket을 사용해서 연결하고 싶고, 서버가 지원한다면 wss://주소
- wss(Secure Web Socket)
- ws(Web Socket)

브라우저가 서버로 webSocket request를 보내면, 서버가 받거나 거절

연결이 되어 있을 경우에만 발생

- 한번 요청을 수락하면, 연결은 성립(establish)
- 연결은 연결, 브라우저와 서버가 손을 맞잡고 있는 것 처럼
- 서버는 유저가 누구인지 알수 있음
- 연결되 있기 때문에 서버가 유저에게 메시지를 보낼수 있음
- 추가적인 request, response 없음
- bi-directional(양 방향의) 연결
- 서로에게 직접 접근할 수 있음
- 서버는 어떤 때나 유저에게 메시지를 보낼수 있음 유저도 마찬가지

- 브라우저와 서버 사이에 bi-directional connection 있어 서로에게 바로 갈수 있는 길이 열림

---

### WebSocket이란 

- Transport protocol의 일종으로 쉽게 이야기하면 웹버전의 TCP 또는 Socket이라고 이해하면 됨

- WebSocket은 서버와 클라이언트 간에 Socket Connection을 유지해서 언제든 양방향 통신 또는 데이터 전송이 가능하도록 하는 기술

- Real-time web application구현을 위해 널리 사용되어지고 있음 
  (SNS어플리케이션, LoL같은 멀티플레이어 게임, 구글 Doc, 증권거래, 화상채팅 등)


### 사용이유

- 웹어플리케이션에서 기존의 서버와 클라이언트 간의 통신은 대부분 HTTP를 통해 이루어 졌으며 
 
  HTTP는 Request/response기반의 Stateless protocol임


- 서버와 클라이언트 간의 Socket connection같은 영구적인 연결이 되어있지 않고 

  클라이언트 쪽에서 필요할때 Request를 할때만 서버가 Response를 하는 방식으로 통신이 진행되는 한방향 통신  
  
  
- 이럴경우 서버쪽 데이터가 업데이트 되더라도 클라이언트 쪽에는 화면은 Refresh하지 않는한 변경된 데이터가 업데이트 되지 않는 문제가 발생


- 이런 문제는 일반적은 웹어플리케이션에서는 기존의 있던 임시방편인 Long polling이라던가 Ajax를 사용해도 어느정도 해결이 가능하지만 
  
  데이터의 빠른 업데이트가 아주 중요한 요소 중에 하나인 어플리케이션에서는 실시간 업데이트가 아주 중요하기 때문에 
  
  Web Socket이 아주 중요한 기술로 사용


- Web Socket은 Stateful protocol이기 때문에 클라이언트와 한 번 연결이 되면 계속 같은 라인을 사용해서 통신하기 때문에 
  
  HTTP 사용시 필요없이 발생되는 HTTP와 TCP연결 트래픽을 피할 수 있음
  
  
- Web Socket은 HTTP와 같은 포트(80)을 사용하기에 기업용 어플리케이션에 적용할 때 방화벽은 재설정 하지 않아도 되는 장점



### 작동원리

- 서버와 클라이언트 간의 WebSocket연결은 HTTP프로토콜을 통해 이루어짐


- 만약 연결이 정상적으로 이루어 진다면 서버와 클라이언트 간에 WebSocket연결이 이루어지고 
  
  일정 시간이 지나면 HTTP연결은 자동으로 끊어짐


### HTTP통신방법과 WebSocket의 차이점

- 결정적인 차이는 프로토콜

- WebSocket 프로토콜은 접속 확립에 HTTP를 사용하지만, 
 
  그 후 통신은 WebSocket 독자의 프로토콜로 이루어짐


- header가 상당히 작아 overhead가 적은 특징이 있음


- 장시간 접속을 전제로 하기 때문에, 접속한 상태라면 클라이언트나 서버로부터 데이터 송신이 가능

  더불어 데이터의 송신과 수신에 각각 커넥션을 맺을 필요가 없어 하나의 커넥션으로 데이터를 송수신 할 수 있음
